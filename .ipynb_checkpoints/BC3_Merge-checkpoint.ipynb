## 3. Data Preparation

### Feature Engineering

```python
#Getting average days between orders as a feature by using days_since_prior_order
orders['average_days_between_orders'] = orders.groupby('user_id')['days_since_prior_order'].mean()
orders['average_days_between_orders'] = orders['average_days_between_orders'].replace(np.nan, 0)
```

```python
#Getting the total items
orders['total_items'] = orders.groupby('user_id').size()
orders['total_items'] = orders['total_items'].replace(np.nan, 0)
```

```python
# getting average basket as a feature by using total items and number of orders
orders['average_basket'] = (orders.total_items /orders.order_number)
orders.head()
```

##### New_Customer

```python
orders.insert(4, 'New_Customer', 0)

for i in range (len(orders)):
    if np.isnan(orders['days_since_prior_order'][i]):
            orders['New_Customer'][i] = 1
    else:
            orders['New_Customer'][i] = 0
orders
```

#### Missing Values - agora substitui por 0, mas podemos pensar na melhor maneira caso usarmos algum modelo

```python
orders['days_since_prior_order'].fillna(0,inplace=True)
```

## Association Rules

```python
df_AR = pd.merge (orders, orderProducts, how='left', on='order_id')
df_AR
```

```python
df_AR = pd.merge(df_AR, products, how='left', on='product_id')
```

```python
basket = (df_AR.groupby(['user_id', 'product_name'])['order_number'].count().unstack().reset_index().fillna(0).set_index('user_id'))
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.8)]
```

```python
rules[(rules['lift'] < 1.1) & (rules['confidence'] < 0.5)]
```

## Recommender System

#### Filtering the data by reorders


```python
# order_products = orderProducts.merge(products, on ='product_id', how='left')
# order_products = order_products.merge(departments, on ='department_id', how='left')
```

```python
reorders = order_products[order_products['reordered'] == 1]
reorders
```

```python
reorders['product_id'] = reorders['product_id'].astype('int64')
```

```python
# get list of high volume products (products that occurr more than once)
hivol = reorders.copy()['product_id'].value_counts().sort_values(ascending=False)\
    [reorders.copy()['product_id'].value_counts().sort_values(ascending=False) > 1].index.tolist()
```

```python
# mask the reorders dataframe to only incluse dem hi vol prods
reorders = reorders[reorders['product_id'].isin(hivol)]
```

```python
#filters the High demand items greater than 1
reorders['high_demand'] = (reorders.copy()['product_id'].value_counts().sort_values(ascending=False)>1)
hidem_ord = reorders[reorders['high_demand'] == True]
```

```python
user_orders = reorders.merge(orders)
```

```python
user_orders['high_demand'] = (user_orders.copy()['product_id'].value_counts().sort_values(ascending=False)>1)
hidem_ord = user_orders[user_orders['high_demand'] == True]
```

#### Setup Model - Compare Users to other Users

```python
#return the total items
users = hidem_ord.groupby(['user_id','product_name']).size().sort_values(ascending=False).unstack().fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 

#creates a similiarity by users.
users_sim = pd.DataFrame(cosine_similarity(users),index=users.index,columns=users.index)
```

```python
def next_prod(df,num_col):
    return df[df.columns[num_col]].drop(df.columns[num_col]).sort_values(ascending=False).head(3)
```

```python
#returns similar users to this one.
pd.DataFrame(next_prod(users_sim,23)).T
```

#### Product Recommender by Order ID

```python
baskets = hidem_ord.groupby(['product_name','order_id']).size().sort_values(ascending=False).unstack().fillna(0)
```

```python
basket_sim = pd.DataFrame(cosine_similarity(baskets),columns=baskets.index,index=baskets.index)
basket_sim
```

```python
basket_sim['baking ingredients'].sort_values(ascending=False).head(10)[1:]
```

```python
def Recommender_System(user_id):
    
    '''
    Input User Id to see recommendations for the User
    '''
    
    u = hidem_ord.groupby(['user_id','product_name']).size().sort_values(ascending=False).unstack().fillna(0)
    u_sim = pd.DataFrame(cosine_similarity(u),index=u.index,columns=u.index)

    p = hidem_ord.groupby(['product_name','user_id']).size().sort_values(ascending=False).unstack().fillna(0)
    products_sim = pd.DataFrame(cosine_similarity(p),index=p.index,columns=p.index)
    profile = pd.Series(np.zeros(len(u_sim.columns.tolist())),u_sim[user_id])
    
    recommendations = pd.Series(np.dot(p.values,users_sim[user_id]), index=p.index)
    return recommendations.sort_values(ascending=False).head()
```
