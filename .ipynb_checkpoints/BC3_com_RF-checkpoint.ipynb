## 3. Data Preparation

### Feature Engineering

```python
#Getting average days between orders as a feature by using days_since_prior_order
orders['average_days_between_orders'] = orders.groupby('user_id')['days_since_prior_order'].mean()
orders['average_days_between_orders'] = orders['average_days_between_orders'].replace(np.nan, 0)
```

```python
#Getting the total items
orders['total_items'] = orders.groupby('user_id').size()
orders['total_items'] = orders['total_items'].replace(np.nan, 0)
```

```python
# getting average basket as a feature by using total items and number of orders
orders['average_basket'] = (orders.total_items /orders.order_number)
orders.head()
```

```python
orders.insert(4, 'New_Customer', 0)

for i in range (len(orders)):
    if np.isnan(orders['days_since_prior_order'][i]):
            orders['New_Customer'][i] = 1
    else:
            orders['New_Customer'][i] = 0
orders
```

```python
orders['days_since_prior_order'].fillna(0,inplace=True)
```

## Association Rules

```python
df_AR = pd.merge (orders, orderProducts, how='left', on='order_id')
df_AR
```

```python
df_AR = pd.merge(df_AR, products, how='left', on='product_id')
```

```python
basket = (df_AR.groupby(['user_id', 'product_name'])['order_number'].count().unstack().reset_index().fillna(0).set_index('user_id'))
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

### Frequent itemsets

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

# We want association rules that have high confidence, high support, and high lift.

#### association rules - by confidence

```python
# Generate the association rules - by confidence
rulesConfidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.750)
rulesConfidence.sort_values(by='confidence', ascending=False, inplace=True)
rulesConfidence.head()
```

### association rules - by lift

```python
# Generate the association rules - by lift
rulesLift = association_rules(frequent_itemsets, metric="lift", min_threshold=1.5)
rulesLift.sort_values(by='lift', ascending=False, inplace=True)
rulesLift.head()
```

```python
rulesLift.describe()
```

### High Confidence and high Lift - complementary products
Sorted by support

```python
# High Confidence and high Lift - complementary products
rulesConfidence[(rulesConfidence['confidence'] >= 0.75) & (rulesConfidence['lift'] >= 1.5)].sort_values(by='support',ascending=False)
```

```python
# High Confidence rules where "fresh vegetables" and "fresh fruits" are in the LHS
rulesConfidence[rulesConfidence['antecedents']=={'fresh vegetables', 'fresh fruits'}]  # Because rules are a "frozenset" object, the order of items is not important
```

```python
# High Confidence rules where "packaged vegetables fruits" is in the RHS
rulesConfidence[['packaged vegetables fruits' in elem for elem in rulesConfidence['consequents']]]
```

### antecedent vs consequents, using lift

```python
# Viewing results for one antecedent lhs item against consequents 
import seaborn as sns
import matplotlib.pyplot as plt

# Replacing the sets with strings
rulesLift['antecedents_'] = rulesLift['antecedents'].apply(lambda a: ','.join(list(a)))
rulesLift['consequents_'] = rulesLift['consequents'].apply(lambda a: ','.join(list(a)))

# Transforming the dataframe of rules into a matrix using the lift metric
pivot = rulesLift.pivot(index = 'antecedents_',columns = 'consequents_', values= 'lift')

# Generating a heatmap
sns.heatmap(pivot, annot = False)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

### network graph

```python
import networkx as nx
### Plot a basic network graph of the top 50 confidence rules
# Create a copy of the rules and transform the frozensets to strings
rulesToPlot = rulesConfidence.copy(deep=True)
rulesToPlot['LHS'] = [','.join(list(x)) for x in rulesToPlot['antecedents']]
rulesToPlot['RHS'] = [','.join(list(x)) for x in rulesToPlot['consequents']]
# Remove duplicate if reversed rules
rulesToPlot['sortedRow'] = [sorted([a,b]) for a,b in zip(rulesToPlot.LHS, rulesToPlot.RHS)]
rulesToPlot['sortedRow'] = rulesToPlot['sortedRow'].astype(str)
rulesToPlot.drop_duplicates(subset=['sortedRow'], inplace=True)
# Plot
rulesToPlot=rulesToPlot[:50]
fig = plt.figure(figsize=(20, 20)) 
G = nx.from_pandas_edgelist(rulesToPlot, 'LHS', 'RHS')  
# Each node is an itemset (either source or target) and the edges connect the itemsets if they consist of a high confidence rule
nx.draw(G, with_labels=True, node_size=30, node_color="red", pos=nx.spring_layout(G))
plt.axis('equal')
plt.show()
#fig.savefig('figure.svg')
# TRY TO PLOT THE DIRECTION OF SOURCE -> TARGET
```

#### Simplified RFM- RF

Nota: secalhar dps é preciso ver aqueles new clients uma vez q neles a R foi substituida por zero

vi as categorias neste site: https://esputnik.com/en/blog/analysis-data-e-marketing-article-2-rfm-segmentation

```python
rf_table = orders.groupby('user_id').agg({'order_number':'max','days_since_prior_order':'mean'}).copy()
```

```python
q = rf_table.quantile(q=[0.25,0.5,0.75])
q
```

```python
def recency_function(x): #using days since...
    if x<q['days_since_prior_order'][0.25]:
        return 1
    elif (x>q['days_since_prior_order'][0.25]) and (x<q['days_since_prior_order'][0.5]):
        return 2
    elif (x>q['days_since_prior_order'][0.5]) and (x<q['days_since_prior_order'][0.75]):
        return 3
    else:
        return 4

rf_table['R_quartile']=rf_table['days_since_prior_order'].apply(lambda x: recency_function(x))


def freq_function(x):# using order_number...
    if x<q['order_number'][0.25]:
        return 1
    elif (x>q['order_number'][0.25]) and (x<q['order_number'][0.5]):
        return 2
    elif (x>q['order_number'][0.5]) and (x<q['order_number'][0.75]):
        return 3
    else:
        return 4

rf_table['F_quartile']=rf_table['order_number'].apply(lambda x: freq_function(x))

rf_table['RF_segment']=rf_table['R_quartile'].astype(str)+rf_table['F_quartile'].astype(str)

rf_table.head()
```

```python
rf_table['RF_segment'].value_counts()
```

```python
# Lost customers: 41,42,43,44
# Leaving VIP: 31,21
# Leaving clients:32,33
# One time clients: 34
# Regular clients: 22,23,24,12,13
# VIP: 11
# New clients:14

map_segments={
    '11': 'VIP',
    '12':'Regular clients' ,
    '13':'Regular clients' ,
    '14':'New clients' ,
    '21':'Leaving VIP' ,
    '22':'Regular clients' ,
    '23':'Regular clients' ,
    '24':'Regular clients' ,
    '31':'Leaving VIP' ,
    '32': 'Leaving clients',
    '33':'Leaving clients' ,
    '34': 'One time clients',
    '41':'Lost customers' ,
    '42': 'Lost customers',
    '43': 'Lost customers',
    '44': 'Lost customers',
}
```

```python
rf_table['Segment']=rf_table['RF_segment'].map(map_segments)
rf_table
```
